```python
import streamlit as st
import duckdb

@st.cache_resource
def get_conn():
    return duckdb.connect("madang.db")

def query(sql, return_type="df"):
    conn = get_conn()
    if return_type == "df":
        return conn.execute(sql).df()
    else:
        return conn.execute(sql).fetchall()
```

```python
st.title("Madang Bookstore (DuckDB)")

menu = st.sidebar.selectbox(
    "메뉴 선택",
    ["Customer", "Book", "Orders"]
)

if menu == "Customer":
    df = query("SELECT * FROM Customer")
    st.dataframe(df)  
```
